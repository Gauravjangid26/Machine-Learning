```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/Users/gauravjangid/Desktop/Data Science/data set/covid_toy.csv")
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df['gender']=lb.fit_transform(df['gender'])
df['cough']=lb.fit_transform(df['cough'])
df['city']=lb.fit_transform(df['city'])
df['has_covid']=lb.fit_transform(df['has_covid'])
```

```python
df=df.dropna()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(columns=['has_covid'])
y=df['has_covid']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
clf=lr.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
#[tp,fp]
#[fn,tn]
```

```python
param_grid = [
    {'penalty':['l1','l2','elasticnet','none'],
    'C' : np.logspace(-4,4,20),
    'solver': ['lbfgs','newton-cg','liblinear','sag','saga'],
    'max_iter'  : [100,1000,2500,5000]
}
]
```

```python
from sklearn.model_selection import GridSearchCV
clf1 = GridSearchCV(lr,param_grid = param_grid, cv = 3, verbose=True,n_jobs=-1)
clf1
```

```python
best_clf = clf1.fit(x,y)
best_clf.best_estimator_
```

```python
print(f'Accuracy - : {best_clf.score(x,y):.3f}')
```

```python
# logisticRegression --->Trarget data ---> Categorical ---> Backend activation function ---> 
# Sigmoid Activation Function ---> Probabilty (0-1)
# threshold_value = 0.5 or 1/2 
# if value>0.5 :
#     return 1 
# else:
#     return 0 
```

```python
from sklearn.ensemble import RandomForestClassifier 
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train , y_train) 
```

```python
y_pred = rf.predict(x_test) 
```

```python
accuracy_score(y_test , y_pred) 
```

```python
from sklearn.linear_model import LogisticRegression 
```

```python
print(f'Accuracy - : {best_clf.score(x,y):.3f}')
```
